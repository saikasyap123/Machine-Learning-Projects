```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_samples = []
train_labels = []
```

```python
for i in range(50):
    younger_persons = randint(13,64)
    train_samples.append(younger_persons)
    train_labels.append(1)
    older_persons = randint(65,100)
    train_samples.append(older_persons)
    train_labels.append(0)
```

```python
for i in range(1000):
    younger_persons = randint(13,64)
    train_samples.append(younger_persons)
    train_labels.append(0)
    older_persons = randint(65,100)
    train_samples.append(older_persons)
    train_labels.append(1)
```

```python
print(train_samples)
```

```python
len(train_samples)
```

```python
print(train_labels)
```

```python
len(train_labels)
```

```python
train_samples = np.array(train_samples)
train_labels = np.array(train_labels)
train_labels, train_samples = shuffle(train_labels, train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
train_samples_new = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
len(train_samples_new)
```

```python
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential()
model.add(Dense(16, input_shape=(1,), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(2, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_samples_new, train_labels, batch_size=10, epochs=20, validation_split=0.1, shuffle=True, verbose=2)
```

```python
test_samples = np.array(train_samples)
test_labels = np.array(train_labels)
```

```python
test_samples_new = scaler.fit_transform(test_samples.reshape(-1,1))
```

```python
preds = model.predict(test_samples_new, batch_size=10, verbose = 0)
```

```python
preds
```

```python
preds_labels = model.predict_classes(test_samples_new, batch_size=10, verbose=0)
```

```python
preds_labels
```

```python
for item in zip(test_labels, preds_labels):
    print('Actual {} predicted {}'.format(item[0], item[1]))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(preds_labels, test_labels)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
plt.title('heatmap')
plt.tight_layout()
plt.show()
```
