```python
from keras.preprocessing.text import one_hot
from keras.models import Sequential
from keras.optimizers import Adam
from keras.metrics import mse
```

```python
words = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'your videos are good']
```

```python
voc_size = 10000
```

```python
one_hot_repr = [one_hot(word, voc_size) for word in words]
```

```python
one_hot_repr
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
same_one_hot = pad_sequences(one_hot_repr, padding='pre',maxlen=8)
```

```python
same_one_hot
```

```python
from keras.layers import Embedding
```

```python
model = Sequential()
model.add(Embedding(voc_size,10,input_length=8))
```

```python
model.compile(Adam(lr=0.0001),metrics=['mse'],loss='sparse_categorical_crossentropy')
```

```python
model.predict(same_one_hot)
```
