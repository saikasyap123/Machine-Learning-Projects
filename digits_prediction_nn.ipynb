```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
import keras 
import keras.backend as k
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.optimizers import Adam
from keras.metrics import sparse_categorical_crossentropy
```

```python
data = pd.read_csv('digits.csv')
data.head()
```

```python
data.isnull().values.any()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='label', data=data)
plt.xlabel('class')
plt.ylabel('count')
plt.title('count plot')
```

```python
features = data.iloc[:,1:].values
target = data.iloc[:,0].values
```

```python
features[0].shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.25)
```

```python
model = Sequential()
model.add(Dense(128, input_dim=784, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(4, activation='softmax'))
```

```python
model.compile(Adam(lr=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=1000, shuffle=True, validation_split=0.15, verbose=2)
```

```python
pred_cls = model.predict_classes(x_test, verbose=0)
```

```python
pred_cls
```

```python
accuracy_score(pred_cls, y_test)
```

```python
cm = confusion_matrix(pred_cls, y_test)
cm
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
plt.title('confusion matrix')
```
